# Problem 1 

1. How many rasters are in the folder?

```python
import os
import arcpy
```

```python

# Set environment settings
arcpy.env.workspace = r"C:\Users\syed_\OneDrive\Documents\ArcGIS\Projects\Project 1\imagery"
```

```python
rasters = arcpy.ListRasters()
num_rasters = len(rasters)
print(f"Number of rasters: {num_rasters}")
```

2. What is the projection of the rasters?

```python
# Get the projection of the first raster
for i in range(num_rasters):
    raster = rasters[i]
    raster_name = arcpy.Describe(raster).baseName
    spatial_ref = arcpy.Describe(raster).spatialReference
    print(f'Raster {raster_name} Projection: {spatial_ref.name}')
```

3. Do all the rasters have the same projection?

Yes. All the rasters have WGS_1984_UTM_Zone_15N as projection. 

4. How many bands to the rasters have?

```python
# Get the number of bands of each raster
for i, raster in enumerate(rasters):
    raster_name = arcpy.Describe(raster).baseName
    band_count = arcpy.Describe(raster).bandCount
    print(f'Number of bands in raster {raster_name}: {band_count}')
```

5. What is the raster format?

```python
# Get the format of each raster

for i, raster in enumerate(rasters):  
    raster_name = arcpy.Describe(raster).baseName
    raster_format = arcpy.Describe(raster).format
    print(f'The format of raster {raster_name} is {raster_format}.')
```

```python
# Loop through the feature classes and get their projections
for feature_class in feature_classes:
    # Use arcpy.Describe to get the spatial reference information
    desc = arcpy.Describe(feature_class)
    # Get the name of the spatial reference (projection)
    projection = desc.spatialReference.name
    print(f"Projection of {feature_class}: {projection}")
```

```python
# Loop through the feature classes and get their shape types
for feature_class in feature_classes:
    # Get the shape type (geometry type)
    shape_type = desc.shapeType
    print(f"Shape type of {feature_class}: {shape_type}")
```

Bonus (5 points) Write your results out to a text file or comma-separated text file.

```python
arcpy.env.overwriteOutput = True

# Open a text file for writing the results
with open("feature_class_info.txt", "w") as file:
    # Write a header line
    file.write("Feature Class Name,Number of Features,Projection,Shape Type\n")

    # Loop through the feature classes and get their information
    for feature_class in feature_classes:
        # Use arcpy.Describe to get the shape type
        desc = arcpy.Describe(feature_class)

        # Get the shape type (geometry type)
        shape_type = desc.shapeType

        # Get the number of features
        count_result = arcpy.GetCount_management(feature_class)
        num_features = int(count_result.getOutput(0))

        # Get the projection
        projection = desc.spatialReference.name if desc.spatialReference else "N/A"

        # Write the information to the text file
        file.write(f"{feature_class},{num_features},{projection},{shape_type}\n")

print("Results have been written to 'feature_class_info.txt'.")
```

```python
# Set environment settings 
arcpy.env.workspace = r"C:\Users\syed_\OneDrive\Documents\ArcGIS\Projects\Project 1\usa_cities.gdb" 
#list feature classes
feature_classes = arcpy.ListFeatureClasses()

#printing feature classes
print("\nFeature Classes:")
for feature_class in feature_classes:
    print(f"- {feature_class}")
```

```python
# Set the name of the feature class
feature_class = "usa_cities"

# Use arcpy.ListFields to list the fields
field_list = arcpy.ListFields(feature_class)

# Print the field names
print("Field Names:")
for field in field_list:
    print(field.name)
```

```python

gdb_path=r"C:\Users\syed_\OneDrive\Documents\ArcGIS\Projects\Project 1\usa_cities.gdb"
# Define the population threshold
population_threshold = 1000000  # 1 million

# Create a search cursor to iterate through the feature class
fields = ["NAME", "POPULATION"]
with arcpy.da.SearchCursor(f"{gdb_path}\\{feature_class}", fields) as cursor:
    # Initialize an empty list to store the results
    results = []

    # Iterate through the rows
    for row in cursor:
        city_name, population = row
        if population > population_threshold:
            results.append((city_name, population))

# Print the results in a table format
print("{:<30} {:<15}".format("City Name", "Population"))
print("-" * 45)
for city, population in results:
    print("{:<30} {:<15,}".format(city, population))
```

# Problem 4

Unzip usa_cities.gdb.zip. Create a list of city capitals and their corresponding state.

In your answer you must use arcpy.da.SearchCursor to iterate through the cities feature class to create a list of the cities and corresponding populations. Not using the search cursor will result in a zero.

```python
import arcpy
gdb_path=r"C:\Users\syed_\OneDrive\Documents\ArcGIS\Projects\Project 1\usa_cities.gdb"

# Set the name of the feature class containing city data
feature_class = "usa_cities"

# Create an empty list to store city capitals and states
city_capitals = []

# Create a feature layer from the input feature class
arcpy.MakeFeatureLayer_management(f"{gdb_path}\\{feature_class}", "cities_layer")

# Use a search cursor to iterate through the feature class
fields = ["NAME", "ST", "CAPITAL"]
with arcpy.da.SearchCursor("cities_layer", fields) as cursor:
    for row in cursor:
        city_name, state_name, is_capital = row
        if is_capital:
            city_capitals.append((city_name, state_name))


# Print the list of city capitals and their corresponding states
for city, state in city_capitals:
    print(f"{city}, {state}")
```
